### 1. ARITHMATIC OPERATORS :

```python
## ADDITION OF NUMBERS  -----
a=10
b=5
c=10.5
d=5.5

add1 = a+b             #whole number
add2 = a+b+c+d         #multiple number
add3 = c+d             #float number
add4 = a+d             #integer + float

print("1)", add1)             # here we are adding serial number to simplify...
print("2)", add2)
print("3)", add3)
print("4)", add4)

print("5)", type(add4))        #the result of additon between integer and float value will always be in float class
```

```python
## SUBSTRACTION OF NUMBERS
a=10
b=5
c=10.5
d=5.5

sub1 = a-b             #whole number
sub2 = a-b-c-d         #multiple number
sub3 = c-d             #float number
sub4 = a-d             #integer - float

print("1)", sub1)
print("2)", sub2)
print("3)", sub3)
print("4)", sub4)

print("5)", type(sub4))      #the result of substraction between integer and float value will always be in float class
```

```python
# here we are going to perform others arithmatic operations :

mul1 = a*b        #MULTIPLICATION
div1 = a/b        #DIVISION
rem1 = a%b        #REMAINDER
pwr  = a**b       #POWER
flr  = a//b       #FLOOR


print("1)", mul1)
print("2)", div1)
print("3)", rem1)
print("4)", pwr)
print("5)", flr)


#In pyhton also "BODMAS" rule apply for the calculations:

result = (3+2)*(8/2)/2                                    # i.e. (5)*(4)/2 -> 20/2 (multiplication first) -> 10
print("6)", result)
```
